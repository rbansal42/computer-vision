```python
# Importing required packages

import cv2
print("Successfully imported")
```

```python
# Reading image

OG_img = cv2.imread("og_img.jpeg")
print("Image read successfully")

# Viewing imported image
print("Rendering Image")
cv2.imshow("img", OG_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Resizing Image
reduced_img = cv2.resize(OG_img, (300,200))
print("Converted Successfully")

# Viewing Resized Image
print("Rendering Image")
cv2.imshow("img", reduced_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Rotating Image
rotated_img = cv2.rotate(OG_img, cv2.ROTATE_180)

# Viewing Rotated Image
cv2.imshow("img", rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Flipping Image
hz_flipped_img = cv2.flip(reduced_img, 0) #Horizontal Flip
vt_flipped_img = cv2.flip(reduced_img, 1) #Vertical Flip

# Viewing Flipped Image
print("Rendering horizontally flipped image")
cv2.imshow("img", hz_flipped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

print("Rendering vertically flipped image")
cv2.imshow("img", vt_flipped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Blurring Image
blurred_img = cv2.blur(reduced_img, (5,5)) #Less Blur
more_blurred_img = cv2.blur(reduced_img, (50,50)) #High Blur

# Viewing Blurred Image
print("Rendering less blurry image")
cv2.imshow("img", blurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

print("Rendering more blurry image")
cv2.imshow("img", more_blurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# GBlurring Image
gBlurred_img = cv2.GaussianBlur(reduced_img, (5,5), 0) #Gausian Blur

# Viewing Gausian Blurred Image
print("Rendering image")
cv2.imshow("img", gBlurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Applying filters to Image
filter1_img = cv2.boxFilter(reduced_img, -1, (5,5)) #Less Blur
filter2_img = cv2.boxFilter(reduced_img, -1, (50,50)) #High Blur

# Viewing Blurred Image
print("Rendering less blurry image")
cv2.imshow("img", filter1_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

print("Rendering more blurry image")
cv2.imshow("img", filter2_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
