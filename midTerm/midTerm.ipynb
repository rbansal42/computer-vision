# Question 1

```python
# Reading image

OG_img = cv2.imread("og_img.jpeg")
print("Image read successfully")

# Viewing imported image
print("Rendering Image")
cv2_imshow(OG_img)
```

```python
# Resizing Image
reduced_img = cv2.resize(OG_img, (300,200))
print("Converted Successfully")

# Viewing Resized Image
print("Rendering Image")
cv2_imshow(reduced_img)
```

```python
# Finding Shape of Image
shape = OG_img.shape
shape1 = reduced_img.shape

# Printing shape of image
print(shape)
print(shape1)
```

# Question 2

```python
# Converting RGB to BGR
bgr_image = cv2.cvtColor(OG_img, cv2.COLOR_RGB2BGR)

# Printing BGR Image
cv2_imshow(cv2.resize(bgr_image, (300,200)))
```

```python
# Cropping Image
cropped_img = OG_img[300:600, 500:800]

# Rendering Image
cv2_imshow(cropped_img)
```

# Question 3

```python
# Finding Edges of image
cannyImg = cv2.Canny(OG_img, threshold1= 100, threshold2= 200)

# Rendering Image
cv2_imshow(cv2.resize(cannyImg, (300,200)))
```

# Question 4

```python
# Blurring Image
blurred_img = cv2.blur(OG_img, (25,25))

# Rendering Image
print("Original Image")
cv2_imshow(cv2.resize(OG_img, (300,200)))
print("Blurred Image")
cv2_imshow(cv2.resize(blurred_img, (300,200)))
```

# Question 5

```python
# Getting a black image
blackImage = cv2.imread("blackCanvas.jpg")
cv2_imshow(blackImage)
```

```python
# Printing shape of image
print(blackImage.shape)
```

```python
# Making a rectangle on the image
rectangle_img = cv2.rectangle(blackImage, (5,5), (195,195), (36, 39, 40), -1)
cv2_imshow(rectangle_img)
```

```python
# Writing text
font = cv2.FONT_HERSHEY_COMPLEX
text_img = cv2.putText(rectangle_img,'text',(70,110),font,1,(244, 245, 245),2,cv2.LINE_AA)

# Rendering Image
cv2_imshow(text_img)
```
