```python
# Importing required packages

import cv2
print("Successfully imported")
```

### Reading Image

```python
# Reading image

OG_img = cv2.imread("og_img.jpeg")
print("Image read successfully")

# Viewing imported image
print("Rendering Image")
cv2.imshow("img", OG_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### Use Different Transformations

```python
# Resizing Image
reduced_img = cv2.resize(OG_img, (300,200))
print("Converted Successfully")

# Viewing Resized Image
print("Rendering Image")
cv2.imshow("Resized Image", reduced_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Rotating Image
rotated_img = cv2.rotate(OG_img, cv2.ROTATE_90_CLOCKWISE)
print("Rotated Successfully")

# Viewing Rotated Image
print("Rendering Image")
cv2.imshow("Rotated Image", reduced_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Flipping Image
hz_flipped_img = cv2.flip(reduced_img, 0) #Horizontal Flip
vt_flipped_img = cv2.flip(reduced_img, 1) #Vertical Flip

# Viewing Flipped Image
print("Rendering horizontally flipped image")
cv2.imshow("Horizontally Flipped Image", hz_flipped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

print("Rendering vertically flipped image")
cv2.imshow("Vertically Flipped Image", vt_flipped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### Adding images with a weight of .8

```python
# Reading 2nd image

OG_img1 = cv2.imread("og_img1.jpeg")
print("Image2 read successfully")

print("Rendering Image 2")
cv2.imshow("img", OG_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# Resizing to equal

resize1 = cv2.resize(OG_img, (500, 500))
resize2 = cv2.resize(OG_img1, (500, 500))

cv2.imshow("img", resize1)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

cv2.imshow("img", resize2)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
createdImage = cv2.addWeighted(resize1, 0.5, resize2, 0.8, 1)
print("Rendering createdImage")

cv2.imshow("img", createdImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### Use Sobel Algorithm on the images

```python
sobelImg = cv2.Sobel(src = OG_img, ddepth = cv2.CV_64F, dx=1, dy=0, ksize = 5)

cv2.imshow("Sobel Algorithm", sobelImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### Find the edges of the image

```python
# Finding Edges of image
cannyImg = cv2.Canny(OG_img, threshold1= 100, threshold2= 200)

# Rendering Image
cv2.imshow("Canny Image", cannyImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### Use smoothening filters on the image

```python
averageBlur = cv2.blur(OG_img1, (5, 5)) 

cv2.imshow("Smoothened Image", averageBlur)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
