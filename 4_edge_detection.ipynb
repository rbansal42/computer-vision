```python
# Importing required packages

import cv2
print("Successfully imported")
```

```python
# Reading image

OG_img = cv2.imread("og_img.jpeg")
print("Image read successfully")

# Viewing imported image
print("Rendering Image")
cv2_imshow(OG_img)
```

```python
sobelX = cv2.Sobel(src = OG_img, ddepth = cv2.CV_64F, dx=1, dy=0, ksize = 5)
sobelY = cv2.Sobel(src = OG_img, ddepth = cv2.CV_64F, dx=0, dy=1, ksize = 5)
sobelZ = cv2.Sobel(src = OG_img, ddepth = cv2.CV_64F, dx=1, dy=1, ksize = 5)
```

```python
cv2_imshow(sobelX)
cv2_imshow(sobelY)
cv2_imshow(sobelZ)
```

```python
cannyImg = cv2.Canny(OG_img, threshold1= 100, threshold2= 200)
cv2_imshow(cannyImg)
```
