```python
# Installing Dependencies

%pip install opencv-python
from google.colab.patches import cv2_imshow
```

```python
# Importing required packages

import cv2
print("Successfully imported")
```

```python
# Reading image

OG_img = cv2.imread("og_img.jpeg")
print("Image1 read successfully")
OG_img1 = cv2.imread("og_img1.jpeg")
print("Image2 read successfully")

# Viewing imported image
print("Rendering Image 1")
cv2_imshow(OG_img)
print("Rendering Image 2")
cv2_imshow(OG_img1)
```

```python
# Resizing to equal

resize1 = cv2.resize(OG_img, (500, 500))
resize2 = cv2.resize(OG_img1, (500, 500))
cv2_imshow(resize1)
cv2_imshow(resize2)
```

```python
createdImage = cv2.add(resize1, resize2, 0)
print("Rendering createdImage")
cv2_imshow(createdImage)
```

```python
createdImage1 = cv2.addWeighted(resize1, 0.3, resize2, 0.5, 1)
cv2_imshow(createdImage1)
```

```python
createdImage2 = cv2.subtract(resize1, resize2)
cv2_imshow(createdImage2)
```

```python
createdImage3 = cv2.multiply(resize1, resize2)
cv2_imshow(createdImage3)
```
